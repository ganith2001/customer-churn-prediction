```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("train.csv")
```

```python
df.drop('area_code',inplace=True,axis=1)
```

```python
df
```

```python
mapping={
    "no":0,
    "yes":1
}
```

```python
df['churn']=df['churn'].map(mapping)
```

```python
df.isnull().values.any()
```

```python
df['churn'].value_counts()
```

```python
df['international_plan']=df['international_plan'].map(mapping)
```

```python
df['voice_mail_plan']=df['voice_mail_plan'].map(mapping)
```

```python
df=df.drop('state',axis=1)
```

```python
df
```

```python
df2=df.drop('churn',axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(df2)
scaled_data=scaler.transform(df2)
scaled_data
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=7)
pca.fit(scaled_data)
x_pca=pca.transform(scaled_data)
dfr=pd.DataFrame(data=x_pca)
dfr
```

```python
print(x_pca.shape)
print(scaled_data.shape)
```

```python

from sklearn.model_selection import train_test_split
import numpy as np
X_train, X_test, y_train, y_test= train_test_split(scaled_data, df['churn'], test_size= 0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_test.value_counts()
```

```python
from sklearn.neighbors import KNeighborsClassifier 
x=[]
y=[]
knn_max=0

l=50
for i in range(1,l+1):
    knn2= KNeighborsClassifier(n_neighbors=i)
    knn2.fit(X_train,y_train)
    s=knn2.score(X_test,y_test)
    if(s>knn_max):
        knn_max=s
        k=i
    x.append(i)
    y.append(s)
```

```python
print(k,knn_max)
plt.plot(x, y)  
plt.title("KNeighborsClassifier")
plt.xlabel("K value")
plt.ylabel("Accuracy")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x=[]
y=[]
dt_max=0

l=30
for i in range(1,l+1):
    dtree = DecisionTreeClassifier(max_depth=i)
    dtree = dtree.fit(X_train, y_train)
    s=dtree.score(X_test,y_test)
    if(s>dt_max):
        dt_max=s
        dt=i
    x.append(i)
    y.append(s)
```

```python
plt.plot(x, y)  
plt.title("DecisionTreeClassifier")
plt.xlabel("depth")
plt.ylabel("Accuracy")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
log_score=logreg.score(X_test,y_test)
log_score
```

```python
from sklearn.ensemble import RandomForestClassifier
x=[]
y=[]
rf_max=0

l=50
for i in range(1,l+1):
    rf_model=RandomForestClassifier(n_estimators=i)
    rf_model = rf_model.fit(X_train, y_train)
    s=rf_model.score(X_test,y_test)
    if(s>rf_max):
        rf_max=s
        rf=i
    x.append(i)
    y.append(s)
print(rf,rf_max)
plt.plot(x, y)
plt.title("RandomForestClassifier")
plt.xlabel("K value")
plt.ylabel("Accuracy")
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
x=[]
y=[]
gb_max=0

l=50
for i in range(1,l+1):
    gbc_model=GradientBoostingClassifier(n_estimators=i)
    gbc = gbc_model.fit(X_train, y_train)
    s=gbc_model.score(X_test,y_test)
    if(s>gb_max):
        gb_max=s
        gb=i
    x.append(i)
    y.append(s)
print(gb,gb_max)
plt.plot(x, y)  
plt.title("GradientBoostingClassifier")
plt.xlabel("K value")
plt.ylabel("Accuracy")
plt.show()
```

```python
x=[]
y=[]
cv_max=0

l=50
for i in range(1,l+1):
    knn= KNeighborsClassifier(n_neighbors=i)
    kf=KFold(n_splits=5)
    knn_score=cross_val_score(knn, X_train, y_train, cv=kf)
    predicted=cross_val_predict(knn, X_test, y_test, cv=kf)
    cv_knn=metrics.accuracy_score(y_test,predicted)
    if(cv_knn>cv_max):
        cv_max=cv_knn
        cv=i
    x.append(i)
    y.append(cv_knn)
print(cv,cv_max)
plt.plot(x, y)  
plt.title("GradientBoostingClassifier")
plt.xlabel("K value")
plt.ylabel("Accuracy")
plt.show()
```

```python
DecisionTree = DecisionTreeClassifier(max_depth=7)
kf=KFold(n_splits=7)
rt_score=cross_val_score(DecisionTree, X_train, y_train, cv=kf)
print('accuracy obtained using Decision Tree: ', rt_score)
predicted=cross_val_predict(DecisionTree, X_test, y_test, cv=kf)
cv_dt=metrics.accuracy_score(y_test,predicted)
print('accuracy: ',cv_dt)
```

```python
RandomForest = RandomForestClassifier(n_estimators=33)
kf=KFold(n_splits=7)
rf_score=cross_val_score(RandomForest, X_train, y_train, cv=kf)
print('accuracy obtained using Regression Tree: ', rf_score)
predicted=cross_val_predict(RandomForest, X_test, y_test, cv=kf)
cv_rf=metrics.accuracy_score(y_test,predicted)
print('accuracy: ',cv_rf)
```

```python
GradientBoosting = GradientBoostingClassifier(n_estimators=36)
kf=KFold(n_splits=7)
rf_score=cross_val_score(GradientBoosting, X_train, y_train, cv=kf)
print('accuracy obtained using Regression Tree: ', rf_score)
predicted=cross_val_predict(GradientBoosting, X_test, y_test, cv=kf)
cv_gb=metrics.accuracy_score(y_test,predicted)
print('accuracy: ',cv_gb)
```
